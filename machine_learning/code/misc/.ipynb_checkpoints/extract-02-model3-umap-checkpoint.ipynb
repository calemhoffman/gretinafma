```python
res_df1 = pd.read_csv('../data_02/F17_alpha_1.csv',header=None)
res_df2 = pd.read_csv('../data_02/F17_alpha_0_2K.csv',header=None)
res_df = pd.concat([res_df1, res_df2])
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(res_df, test_size=0.5, random_state=42)
X_train = train.iloc[:,range(1,36)].values
y_train = train.iloc[:,37].values.reshape(-1, 1)
X_test = test.iloc[:,range(1,36)].values
y_test = test.iloc[:,37].values.reshape(-1, 1)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
X_preproc = Pipeline([('stdscaler', StandardScaler()), ('minmax', MinMaxScaler(feature_range=(0, 1)))])
y_preproc = Pipeline([('onehot',OneHotEncoder(sparse=False))])
```

```python
X_train_preproc = X_preproc.fit_transform(X_train)
X_test_preproc = X_preproc.transform(X_test)
y_train_preproc = y_preproc.fit_transform(y_train)
y_test_preproc = y_preproc.transform(y_test)
```

```python
print(X_train_preproc.shape)
print(X_test_preproc.shape)
print(y_train_preproc.shape)
print(y_test_preproc.shape)
```

```python
print(y_train_preproc)
```

```python
pd.DataFrame(y_train).hist()
pd.DataFrame(y_test).hist()
```

```python
import numpy as np
X_data = np.concatenate((X_train_preproc,X_test_preproc))
y_data = np.concatenate((y_train,y_test))
print(X_data.shape)
print(y_data.shape)
```

```python
import umap
```

```python
np.random.seed(42)
X_trans = umap.UMAP(n_neighbors=30, n_components=2, random_state=42).fit(X_data)
X_model = X_trans.fit(X_data)
X_embedded = X_model.transform(X_data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df_subset = pd.DataFrame()
df_subset['dimension1'] = X_embedded[:,0]
df_subset['dimension2'] = X_embedded[:,1]
df_subset['y'] = y_data
plt.figure(figsize=(16,10))
sns.scatterplot(
    x="dimension1", y="dimension2",
    hue="y",
    palette=sns.color_palette("RdBu", 13),
    data=df_subset,
    legend="full",
    alpha=0.5
)
```

```python
filename = '../data_02/F17_alpha_0_2K.csv'
```

```python
chunksize = 10 ** 6
res_df = pd.DataFrame()
cntr = 0
for chunk in pd.read_csv(filename, chunksize=chunksize, header=None):
    res_df = chunk
    #print(res_df.shape)
    #print(res_df.iloc[:,range(1,36)])
    X_data = res_df.iloc[:,range(1,36)].values
    y_data = res_df.iloc[:,37].values.reshape(-1, 1)
    X_data_preproc = X_preproc.transform(X_data)
    X_embedded = X_model.transform(X_data_preproc)
    ids = res_df.iloc[:,0].values.reshape(-1, 1)
    req_data = pd.concat([pd.DataFrame(ids), pd.DataFrame(y_data), pd.DataFrame(X_embedded)],axis=1)
    req_data.to_csv('../results_02/F17_alpha_umap_%05d.csv'%cntr,header=None,index=False)
    cntr = cntr + 1
    #break
```
